# EXERCISE 2: semi-super resolution!

Take one image on your own (or one blurred image from somewhere) and apply the techniques we have seen in class to sharpen an image from a blurred one. You can also blur one image of your own as a start.

```python
#Load the image 
img_sf = imageio.imread('BasilicaOfTheHolyFamily.jpeg')

# Plotting
fig, ax = plt.subplots(nrows = 1, ncols=1, sharex=False, sharey=False, figsize=(8, 8))

ax.imshow(img_sf)
ax.set_title('SF_Original')

plt.tight_layout()
plt.show;

#plt.imshow(img, cmap=plt.cm.gray)
print("shape of image: {}".format(img_sf.shape))
#uint8 --> 0 to 255
print("dtype of image: {}".format(img_sf.dtype))
#print(img_desc)
#uint8 --> 0 to 255
```

```python
#Sharpening 
#Method 1: Using this method: http://scipy-lectures.org/advanced/image_processing/#sharpening
# Image to grayscale
img_sf_gs = color.rgb2gray(img_sf)

# First we blur the image using a gaussian filter
img_sf_blurred = ndimage.gaussian_filter(img_sf_gs, 3)

# Then we create another filter of the blurred image using sigma 1
img_sf_blurred_filter = ndimage.gaussian_filter(img_sf_blurred, 1)

# We use an approximation of the Laplacian for increasing the weights of edges
alpha = 30
img_sf_sharpened = img_sf_blurred + alpha * (img_sf_blurred - img_sf_blurred_filter)


# Plotting
fig, ax = plt.subplots(nrows = 2, ncols=2, sharex=False, sharey=False, figsize=(15, 15))

ax[0,0].imshow(img_sf_gs, cmap=plt.cm.gray)
ax[0,0].set_title('img_sf')

ax[0,1].imshow(img_sf_blurred, cmap=plt.cm.gray)
ax[0,1].set_title('img_sf_blurred')

ax[1,0].imshow(img_sf_blurred_filter, cmap=plt.cm.gray)
ax[1,0].set_title('img_sf_blurred_filter')

ax[1,1].imshow(img_sf_sharpened, cmap=plt.cm.gray)
ax[1,1].set_title('img_sf_sharpened')

fig.suptitle('Sharpening - Method 1: Laplacian approximation', fontsize=20, y=0.94)

#plt.tight_layout()
plt.show()
```

```python
#
img_sf_blurred_filter_convolution = ReadyMadeKernel_img_RGB(image = img_sf_gs, sigma_est = None, ready_made_kernel = ReadyMadeKernel.BLUR)
img_sf_sharpen_filter_convolution = ReadyMadeKernel_img_RGB(image = img_sf_blurred_filter_convolution, sigma_est = None, ready_made_kernel = ReadyMadeKernel.SHARPEN)

# Plotting
fig, ax = plt.subplots(nrows = 1, ncols=3, sharex=False, sharey=False, figsize=(15, 15))

ax[0].imshow(img_sf_gs, cmap=plt.cm.gray)
ax[0].set_title('img_sf')

ax[1].imshow(img_sf_blurred_filter_convolution, cmap=plt.cm.gray)
ax[1].set_title('img_sf_blurred_filter')

ax[2].imshow(img_sf_sharpen_filter_convolution, cmap=plt.cm.gray)
ax[2].set_title('img_sf_sharpened')

fig.suptitle('Sharpening - Method 2: convolution matrix', fontsize=20, y=0.82)

#plt.tight_layout()
plt.show()
```
